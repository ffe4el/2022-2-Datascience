## 다수의 음원차트 파일 이용하기

```python
# 파일이름 목록 만들기
import os
filenames = [x for x in os.listdir('./weekly_chart/') if x.endswith('.txt')]
filenames
```

```python
# 모든 파일 읽어서 하나의 dataframe으로 만들기
import pandas as pd

top100_list = []
for filename in filenames :
    top100 = pd.read_csv("./weekly_chart/"+filename, sep="|", encoding='utf-8')
    top100['year'] = int(filename[:4])
    top100_list.append(top100)

df = pd.concat(top100_list)
df
```

```python
df.shape
```

```python
# 점수 매기기
df['score']=101-df['rank']
```

```python
df
```

```python
# 전 시기에서 점수대로 순서 세우기
df_sum = df[["artist_name", "score"]].groupby('artist_name').sum()
df_sum = df_sum.sort_values('score', ascending=False)
df_sum = df_sum.reset_index
df_sum
```

```python
# 시기 추가하기
# 2000-2007: 2000년대
# 2008-2015: 2010년대
# 2016-2022: 2020년대

def sep_era(x):
    if x < 2008:
        return '2000년대'
    elif x < 2016:
        return '2010년대'
    return '2020년대'

df['era'] = df['year'].apply(sep_era)
df
```

```python
# 시기 추가하기 lambda버전(lambda 매개변수: 표현식)
# df.apply( lambda x : print(x) )
df['eral']=df['year'].apply(lambda x : '2000년대' if x < 2008 else ('2010년대' if x < 2016 else ('2020년대')))
df
```

```python
# 시기별 인기 가수 찾아보기 (top5) 2000년대
df_era1 = df[df["era"]=='2000년대'].groupby("artist_name").sum()['score'].sort_values(ascending=False)
df_era1.head()
```

```python
# 시기별 인기 가수 찾아보기 (top5) 2010년대
df_era2 = df[df["era"]=='2010년대'].groupby("artist_name").sum()['score'].sort_values(ascending=False)
df_era2.head()
```

```python
# 시기별 인기 가수 찾아보기 (top5) 2020년대
df_era3 = df[df["era"]=='2020년대'].groupby("artist_name").sum()['score'].sort_values(ascending=False)
df_era3.head()
```

```python
# 한번에 시대별 탑5 보기
df_era = df.groupby(["era", 'artist_name']).sum()["score"].sort_values(ascending=False)
df_era = df_era.groupby(['era'])
df_era = df_era.head(5).reset_index().sort_values(['era', 'score'], ascending=[True, False])
df_era
```

## 월드컵 역대 골기록

```python
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 주어진 전체 기간의 각 나라별 골득점수 상위 5개 국가와 그 득점수를 그래프로 출력하라
df2 = df.groupby('Country')['Goals'].sum().sort_values(ascending=False).head()
df2 = df2.reset_index()
display(df2)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(data=df2, x='Country', y='Goals')
plt.show()
```

```python
# Years 컬럼은 년도 -년도 형식으로 구성되어있고, 각 년도는 4자리 숫자이다. 년도 표기가 4자리 숫자로 안된 케이스가 존재한다. 해당 건은 몇건인지 출력하라
def checkFour(x):
    for value in x:
        if len(str(value)) != 4:
            return False
    return True

df['yearLst'] = df['Years'].str.split("-")
df['check']=df['yearLst'].apply(checkFour)
df_clean = df[df['check']].copy()
df_clean
```

```python
df2 = df_clean
df2
```

```python
# 월드컵 출전횟수를 나타내는 ‘LenCup’ 컬럼을 추가하고 4회 출전한 선수의 숫자를 구하여라
# len(df["yearLst"])
# df2['LenCup']=df[len(df["yearLst"])==4].sum()
df2['Lencup']= df2['yearLst'].str.len()
df2['Lencup'].value_counts()[4]
```

```python
# 우리나라의 월드컵 출전횟수가 3회 이상인 선수들의 숫자를 구하여라
# df2['Country'].unique()
df2[df2['Country']=='South Korea'][df2['Lencup']>=3]
```

```python
# 우리나라 선수 중 가장 골을 많이 넣은 선수는?
df2[df2['Country']=='South Korea'].sort_values('Goals', ascending=False).head()
```

```python
# 2002년도에 출전한 전체 선수는 몇명인가?
len(df2[df2['Years'].str.contains('2002')])
```

```python
# 이름에 ‘carlos’ 단어가 들어가는 선수의 숫자는 몇 명인가? (대, 소문자 구분 x)
len(df2[df2['Player'].str.contains('carlos', case=False, regex=False)])
```

```python
# 월드컵 출전 횟수가 1회뿐인 선수들 중에서 가장 많은 득점을 올렸던 선수는 누구인가?
df2[df2['Lencup']==1].sort_values('Goals', ascending=False).head(1)
```

```python
# 월드컵 출전횟수가 1회 뿐인 선수들이 가장 많은 국가는 어디인가?
df2[df2['Lencup']==1].groupby('Country').sum().sort_values('Lencup', ascending=False)
```
