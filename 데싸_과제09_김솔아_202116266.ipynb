```python
# 파일이름 목록 만들기
import os 
filenames = [x for x in os.listdir("./weekly_chart/") if x.endswith('.txt')]
```

```python
# 모든 파일 읽어서 하나의 dataframe으로 만들기
import pandas as pd
df_list = []
for filename in filenames : 
  df_top100_partial = pd.read_csv("./weekly_chart/"+filename, sep = '|', encoding = 'utf-8')
  df_top100_partial['year']= int(filename[:4])
  df_list.append(df_top100_partial)
df = pd.concat(df_list)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["score"]=101-df['rank']
```

```python
df_sum = df[['artist_name', "score"]].groupby('artist_name').sum()
df_sum.sort_values("score", ascending=False).head()
```

```python
# 시기 추가하기
# 2000-2007: 2000년대
# 2008-2015: 2010년대
# 2016-2022: 2020년대

# 함수로 만들기 
def determine_era(x):
  if x <=2007:
    return "2000년대"
  elif 2008 <=x<=2015:
    return "2010년대"
  return "2020년대"
df["시기"] = df["year"].apply(determine_era)

# lambda로 만들기
# df['시기'] = df['year'].apply(lambda x :"2000년대" if x < 2008 else "2010년대")
# df['시기'] = df['year'].apply(lambda x :"2010년대" if x < 2016 else "2020년대")

df.head()
```

```python
# 시기별 인기 가수 찾아보기
df_era = df.groupby(["시기", 'artist_name']).sum()["score"]
df_era
```

```python
# 시기별 top 5 가수 찾기
# https://cmdlinetips.com/2019/03/how-to-get-top-n-rows-with-in-each-group-in-pandas/
df_era = df.groupby(["시기", 'artist_name']).sum()["score"].sort_values(ascending=False)
df_era = df_era.groupby(['시기'])
df_era = df_era.head(5).reset_index().sort_values(['시기', 'score'], ascending=[True, False])
df_era
# df_era2 = df_era.sort_values(['score'], ascending=False)
# df_era2 = df_era2.reset_index()
```

# 예제2. 월드컵 역대 골기록

* 데이터 출처 :https://www.kaggle.com/darinhawley/fifa-world-cup-goalscorers-19302018(참고, 데이터 수정)
* 데이터 설명 : 1930 ~2018년도 월드컵 출전선수 골기록
* data url = https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv
* 관련 글: https://www.datamanim.com/dataset/03_dataq/pandas_practice.html

```python
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
# 주어진 전체 기간의 각 나라별 골득점수 상위 5개 국가와 그 득점수를 그래프로 출력하라
df.groupby('Country')['Goals'].sum().sort_values(ascending=False).head().plot.barh() #bar를 옆으로 그려줌(horizen), 근데 정렬이 거꾸로 되어있음
plt.gca().invert_yaxis() # y축 정렬을 다시 반대로 돌려줌
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 주어진 전체 기간의 각 나라별 골득점수 상위 5개 국가와 그 득점수를 그래프로 출력하라
df2 = df.groupby('Country')['Goals'].sum().sort_values(ascending=False).head()
df2 = df2.reset_index()

fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(data=df2, x='Goals', y='Country')
plt.show()
```

```python
# Years 컬럼은 년도 -년도 형식으로 구성되어있고, 각 년도는 4자리 숫자이다. 년도 표기가 4자리 숫자로 안된 케이스가 존재한다. 해당 건은 몇건인지 출력하라
def checkFour(x):
    for value in x:
        if len(str(value)) != 4:
            return False
    return True

df['yearLst'] = df['Years'].str.split('-')
df['check'] = df['yearLst'].apply(checkFour)
#[df['check'] == False] 아래랑 같은 의미
df[~df['check']]
# df_clean = df[df['check'] == True].copy() 아래랑 같은 의미
df_clean = df[df['check']].copy()
df_clean

```

```python
# 깜짝퀴즈

# 위 함수식을 람다식으로 써보시오.
# check_four_lambda = lambda x : ?
# 답을 작성한 학생은 이메일로 제출하기 바랍니다.

check_four_lambda = lambda x : all([len(str(y)) == 4 for y in x])
df['check'] = df['yearLst'].apply(checkFour)
df['check2'] = df['yearLst'].apply(check_four_lambda)



len(df[df['check']!=df["check2"]]) == 0
```

```python
# 위에서 발생한 예외 케이스를 제외한 데이터프레임을 df2라고 정의하고 데이터의 행의 숫자를 출력하라 (아래 문제부터는 df2로 풀이하겠습니다)

df2 = df[df['check']].copy()
df2
```

```python
# 월드컵 출전횟수를 나타내는 ‘LenCup’ 컬럼을 추가하고 4회 출전한 선수의 숫자를 구하여라

df2['Lencup']= df2['yearLst'].str.len()
df2['Lencup'].value_counts()[4] # 4회 출전한 선수의 숫자

# df2['LenCup'] =df2['yearLst'].str.len()
# result = df2['LenCup'].value_counts()[4]
# print(result)
```

```python
# 우리나라의 월드컵 출전횟수가 3회 이상인 선수들의 숫자를 구하여라
df2[(df2['Country']=='South Korea')&(df2['Lencup']>=3)]
```

```python
# 우리나라 선수 중 가장 골을 많이 넣은 선수는?
df2[df2['Country']=='South Korea'].sort_values('Goals', ascending=False).head()
```

```python
# 이름에 ‘carlos’ 단어가 들어가는 선수의 숫자는 몇 명인가? (대, 소문자 구분 x)
len(df2[df2['Player'].str.contains('carlos', case=False, regex=False)])
```

```python
# 월드컵 출전 횟수가 1회뿐인 선수들 중에서 가장 많은 득점을 올렸던 선수는 누구인가?
df2[df2['Lencup']==1].sort_values('Goals', ascending=False).head(1)
```

```python
# 월드컵 출전횟수가 1회 뿐인 선수들이 가장 많은 국가는 어디인가?
df2[df2['Lencup']==1].groupby('Country').sum().sort_values('Lencup', ascending=False)
```
